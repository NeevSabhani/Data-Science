Explain what happens when we do matrix multiplication.

Yes, I know I basically already asked you to do this in the notes. Don't copy and paste your anwser from before; honestly do the exercise again.

When multiplying two matrices, we multiply the rows of the first matrix with the columns of the second matrix. For example, each number in the first row of the first matrix is multiplied by the corresponding number in the first column of the second matrix. The products are then added together to form a new number, which will go in the top-left position of the new matrix. This process is repeated for each row and column combination, filling in the entire new matrix.
